```python
from tensorflow.keras.models import load_model
import numpy as np
```

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt

img = plt.imread("test_face2.jpg")
img = img
plt.imshow(img, cmap='gray')
```

```python
img = img/255.0
img.shape
```

```python
import cv2

test_img = cv2.resize(img , (224, 224), interpolation=cv2.INTER_CUBIC)
```

```python
test_img.shape
```

```python
test_img = np.expand_dims(test_img, axis=0)
```

```python
test_img.shape
```

```python
model = load_model("densenet121_detection_model.h5")
```

```python
pred = model.predict(test_img)
```
